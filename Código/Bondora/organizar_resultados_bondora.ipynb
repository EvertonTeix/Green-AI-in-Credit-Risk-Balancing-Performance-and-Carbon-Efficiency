```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def plotar_comparativo_final_en(df_red, df_aum):
    models = ['RNA', 'LightGBM', 'RandomForest']
    colors = {'RNA': '#1F77B4', 'LightGBM': '#2CA02C', 'RandomForest': '#D62728'}

    sns.set_theme(style="whitegrid", font="sans-serif")
    plt.figure(figsize=(14, 8), dpi=120)

    def prepare_data(df, type_tag):
        temp = df[df['Modelo'].isin(models)].copy()
        scenario_col = temp.columns[1]

        temp['X_Num'] = temp[scenario_col].astype(str).str.replace('%', '').str.strip()
        temp['X_Num'] = pd.to_numeric(temp['X_Num'], errors='coerce')

        if type_tag == 'reduction':
            temp['X_Num'] = -temp['X_Num'].abs()
        else:
            temp['X_Num'] = temp['X_Num'].abs()

        return temp[['Modelo', 'X_Num', 'RGA']]

    df_plot = pd.concat([prepare_data(df_red, 'reduction'),
                         prepare_data(df_aum, 'increase')])
    df_plot = df_plot.drop_duplicates(subset=['Modelo', 'X_Num']).sort_values('X_Num')

    for m in models:
        subset = df_plot[df_plot['Modelo'] == m]
        if not subset.empty:
            plt.plot(subset['X_Num'], subset['RGA'],
                     marker='o', markersize=9,
                     color=colors[m], label=m,
                     linewidth=3.5, alpha=0.85,
                     markeredgecolor='white', markeredgewidth=1.5)

    plt.yscale('log')
    plt.axvline(x=0, color='#333333', linestyle='--', linewidth=2, alpha=0.7)

    ticks = sorted(df_plot['X_Num'].unique())
    labels = []
    for x in ticks:
        if x > 0: labels.append(f"+{int(x)}%")
        elif x < 0: labels.append(f"{int(x)}%")
        else: labels.append("Baseline (0%)")

    plt.xticks(ticks, labels, rotation=45, fontsize=11)
    plt.yticks(fontsize=11)
    plt.xlabel(r'Feature Variability',
               fontsize=13, fontweight='bold', labelpad=15)
    plt.ylabel('RGA (Emissions per Accuracy) - Measured via CarbonTracker', fontsize=13, fontweight='bold', labelpad=15)
    plt.title('Green AI Sensitivity Analysis: Model Efficiency Comparison\n(Bondora Dataset)',
              fontsize=16, fontweight='bold', pad=25)

    plt.grid(True, which="both", ls="-", alpha=0.3)

    plt.legend(title="ML Models", title_fontsize='12', fontsize='11',
               frameon=True, facecolor='white', edgecolor='0.8',
               bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)

    plt.tight_layout()
    plt.show()

plotar_comparativo_final_en(reducao_carbontracker, aumento_carbontracker)
```
